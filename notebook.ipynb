## Visualization & Explanatory Analysis

### Pertanyaan 1: Apakah cuaca berpengaruh terhadap jumlah pengguna rental sepeda?


```python
# Pengaruh Cuaca Terhadap Jumlah Pengguna Rental Sepeda
plt.figure(figsize=(8, 6))
sns.boxplot(x='weathersit', y='cnt', data=df)
plt.title('Pengaruh Cuaca Terhadap Jumlah Pengguna Rental Sepeda')
plt.xlabel('Situasi Cuaca (1 = Cerah, 2 = Berkabut, 3 = Hujan)')
plt.ylabel('Jumlah Penyewaan Sepeda')
plt.show()



# Pengaruh Cuaca Terhadap Jumlah Pengguna Rental Sepeda dengan Plotly
fig = px.box(df, x='weathersit', y='cnt',
             labels={
                 'weathersit': 'Situasi Cuaca',
                 'cnt': 'Jumlah Penyewaan Sepeda'
             },
             title='Pengaruh Cuaca Terhadap Jumlah Pengguna Rental Sepeda',
             category_orders={"weathersit": [1, 2, 3]},  # Menentukan urutan kategori
             hover_data=['temp', 'hum', 'windspeed'])  # Menampilkan info tambahan saat hover

# Mengatur label kategori cuaca agar lebih deskriptif
fig.update_xaxes(tickvals=[1, 2, 3],
                 ticktext=['Cerah/Mendung', 'Berkabut', 'Hujan Ringan'])

# Menampilkan grafik interaktif
fig.show()
```

**Insight**: Berdasarkan analisis yang dilakukan terhadap pengaruh cuaca terhadap jumlah pengguna rental sepeda, berikut adalah kesimpulan yang dapat diambil:

Perbedaan Jumlah Penyewaan Berdasarkan Situasi Cuaca: Grafik boxplot menunjukkan bahwa terdapat perbedaan signifikan dalam jumlah penyewaan sepeda berdasarkan situasi cuaca.

Cuaca Cerah: Situasi cuaca yang cerah (kode 1) menunjukkan jumlah penyewaan tertinggi, dengan rentang yang lebih besar dan mediana yang jelas di atas kategori lainnya. Ini menunjukkan bahwa banyak pengguna memilih untuk menyewa sepeda saat cuaca baik.
Cuaca Berkabut: Situasi berkabut (kode 2) memperlihatkan penurunan yang signifikan dalam jumlah penyewaan. Ini mungkin menunjukkan bahwa berkabut dapat mengurangi visibilitas, sehingga pengguna lebih memilih untuk tidak menyewa sepeda.
Cuaca Hujan: Situasi hujan (kode 3) menunjukkan jumlah penyewaan terendah. Hujan sering kali menjadi faktor penghambat utama bagi pengguna sepeda, yang mungkin mencari alternatif transportasi lain saat cuaca buruk.
Korelasi Negatif pada Cuaca Buruk: Dari analisis tersebut, terlihat bahwa cuaca buruk, baik berkabut maupun hujan, berdampak negatif terhadap minat masyarakat untuk menggunakan rental sepeda. Pengguna cenderung lebih memilih tidak menyewa sepeda dalam kondisi cuaca yang tidak mendukung.

### Pertanyaan 2: Apakah terdapat perbedaan antara rata-rata jumlah pengguna rental sepeda di weekday dan Weekend?

```python
df['is_weekend'] = np.where(df['weekday'].isin([0, 6]), 'Weekend', 'Weekday')

plt.figure(figsize=(8, 6))
sns.barplot(x='is_weekend', y='cnt', data=df, estimator=np.mean)
plt.title('Rata-rata Jumlah Penyewaan Sepeda di Weekday vs Weekend')
plt.xlabel('Kategori Hari')
plt.ylabel('Rata-rata Jumlah Penyewaan Sepeda')
plt.show()

# Membuat kategori 'Hari Kerja' dan 'Akhir Pekan' menggunakan np.where
df['is_weekend'] = np.where(df['weekday'].isin([0, 6]), 'Weekend', 'Weekday')

# Rata-rata Jumlah Penyewaan di Weekday vs Weekend dengan Plotly
fig = px.bar(df, x='is_weekend', y='cnt',
             labels={
                 'is_weekend': 'Kategori Hari',
                 'cnt': 'Jumlah Penyewaan Sepeda'
             },
             title='Rata-rata Jumlah Penyewaan Sepeda di Weekday vs Weekend',
             hover_data=['temp', 'hum', 'windspeed'],  # Menampilkan informasi tambahan saat hover
             color='is_weekend',  # Pewarnaan berdasarkan kategori (Weekday/Weekend)
             barmode='group',
             category_orders={"is_weekend": ['Weekday', 'Weekend']})  # Mengatur urutan kategori

# Menampilkan grafik interaktif
fig.show()
```

**Insight:**
Berdasarkan analisis yang dilakukan, ditemukan bahwa rata-rata jumlah pengguna rental sepeda di hari kerja (weekday) lebih tinggi dibandingkan dengan akhir pekan (weekend). Hal ini menunjukkan bahwa orang lebih cenderung menggunakan sepeda untuk keperluan sehari-hari, seperti perjalanan ke tempat kerja atau aktivitas rutin lainnya pada hari kerja.

Temuan ini bisa menjadi insight yang penting bagi pengelola layanan rental sepeda dalam merencanakan strategi pemasaran dan penyediaan sepeda, serta mempertimbangkan penawaran khusus atau promosi untuk akhir pekan guna meningkatkan jumlah penyewaan.
